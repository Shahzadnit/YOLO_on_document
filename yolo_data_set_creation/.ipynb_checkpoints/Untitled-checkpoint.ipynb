```python
import os
import json
file2=open("anotation_demo.txt","w")
# List all files in a directory using os.listdir
basepath = "./file1(copy)"
for entry in os.listdir(basepath):
    with open(os.path.join(basepath, entry)) as f:
        data = json.load(f)
        file2.write(entry[0:12])
        
        for i in range(len(data['objects'])):
            file2.write(" ")
            file2.write("%d "%data['objects'][i]['points']['exterior'][0][0])
            file2.write("%d "%data['objects'][i]['points']['exterior'][0][1])
            file2.write("%d "%data['objects'][i]['points']['exterior'][1][0])
            file2.write("%d "%data['objects'][i]['points']['exterior'][1][1])
            clas=data['objects'][i]['classTitle']
            if clas=="Fish":
                file2.write("0")
            if clas=="Rock":
                file2.write("1")
#             file2.write(" ")
    file2.write("\n")
#         print(data)
file2.close()     
    
#     if os.path.isfile(os.path.join(basepath, entry)):
#         print(entry)
```

```python
import os
import cv2
import json
file2=open("anotation_new.txt","w")
# List all files in a directory using os.listdir
basepath = "./ann"
Imagepath = "./img"
Imagesavepath = "./img_new"
arr = os.listdir(Imagepath)
entry = os.listdir(basepath)
j=100000
j1=0
for entry1 in os.listdir(Imagepath):
    img=cv2.imread(os.path.join(Imagepath, entry1))
#     print(img.shape)
    cv2.imwrite(os.path.join(Imagesavepath, "%d.jpg"% (j+1)),img)
    print(entry1)
    print(entry[j1])
    with open(os.path.join(basepath, entry[j1])) as f:
        data = json.load(f)
    file2.write("%d.jpg"% (j+1))
#     file2.write(" ")
    for i in range(len(data['objects'])):
        file2.write(" ")
        file2.write("%d "%data['objects'][i]['points']['exterior'][0][0])
        file2.write("%d "%data['objects'][i]['points']['exterior'][0][1])
        file2.write("%d "%data['objects'][i]['points']['exterior'][1][0])
        file2.write("%d "%data['objects'][i]['points']['exterior'][1][1])
        clas=data['objects'][i]['classTitle']
        if clas=="comp_name":
            file2.write("0")
        if clas=="Total_ammount":
            file2.write("1")
        if clas=="Invoice_date":
            file2.write("2")
        if clas=="Invoice_no":
            file2.write("3")
        if clas=="Due_date":
            file2.write("4")
#         file2.write(" ")
    file2.write("\n")
    j=j+1
    j1=j1+1
#     if(j==9):
#         break
file2.close()     
```
